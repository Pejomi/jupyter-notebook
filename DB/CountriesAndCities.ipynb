```python
import pandas as pd
```

```python
df = pd.read_csv("../DB/Data/combined_emission_report.csv")
df2 = pd.read_csv("../DB/Data/combined_reductions_report.csv")
concat = pd.concat([df, df2], axis=0)
concat
```

```python
concat = concat.dropna(subset=['Country'])
concat.reset_index(drop=True, inplace=True)
```

```python
unique_countries = concat['Country'].unique()
unique_countries
```

```python
unique_cities = concat['City'].unique()
unique_cities
```

```python
unique_countries = concat.drop_duplicates(subset=['Country'])
unique_countries.reset_index(drop=True, inplace=True)
unique_countries[['Country', 'Country Location']].to_csv('../DB/Data/countries.csv', index=False)
```

```python
unique_cities = concat.drop_duplicates(subset=['City'])
unique_cities.reset_index(drop=True, inplace=True)
unique_cities[['Country', 'City', 'City Location', 'C40']].to_csv('../DB/Data/cities.csv', index=False)
```
