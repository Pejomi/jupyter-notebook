```python
import pandas as pd
```

```python
reduction2016 = pd.read_csv("../DB/Data/2016_-_Cities_Emissions_Reduction_Targets_20240207.csv")
reduction2017 = pd.read_csv("../DB/Data/2017_-_Cities_Emissions_Reduction_Targets_20240207.csv")
emission2016 = pd.read_csv("../DB/Data/2016_-_Citywide_GHG_Emissions_20240207.csv")
emission2017 = pd.read_csv("../DB/Data/2017_-_Cities_Community_Wide_Emissions.csv")
report2023 = pd.read_csv("../DB/Data/2023_Cities_Climate_Risk_and_Vulnerability_Assessments_20240207.csv")
```

```python
def check_null(df):
    # Check for null values in each column
    null_counts = df.isnull().sum()

    # Print columns with null values
    print("Columns with null values:")
    print(null_counts[null_counts > 0])
```

```python
# Function to convert C40 to True and NaN to False
def convert_to_boolean(value):
    if value == 'C40':
        return True
    else:
        return False
```

## Reduction 2016

```python
reduction2016
```

```python
check_null(reduction2016)
```

```python
reduction2016.rename(columns={'City Short Name': 'City Name'}, inplace=True)
reduction2016['C40'] = reduction2016['C40'].apply(lambda x: convert_to_boolean(x))

reduction2016.drop(columns=['Organisation'], inplace=True)

reduction2016.head()
```

```python
# Save the DataFrame to a CSV file
reduction2016.to_csv('../DB/Data/new_2016_reductions.csv', index=False)
```
