```python
import pandas as pd
```

```python
def check_null(df):
    # Check for null values in each column
    null_counts = df.isnull().sum()

    # Print columns with null values
    print("Columns with null values:")
    print(null_counts[null_counts > 0])
```

```python
# Function to convert C40 to True and NaN to False
def convert_to_boolean(value):
    if value == 'C40':
        return True
    else:
        return False
```

# 2016

```python
df = pd.read_csv("../DB/Data/2016_-_Cities_Emissions_Reduction_Targets_20240207.csv")
```

```python
df
```

```python
check_null(df)
```

```python
df['C40'] = df['C40'].apply(lambda x: convert_to_boolean(x))
```

```python
df.drop(columns=['Organisation'], inplace=True)
```

```python
df.rename(columns={'Baseline emissions (metric tonnes CO2e)': 'Baseline emissions'}, inplace=True)
```

```python
df.rename(columns={'City Short Name': 'City'}, inplace=True)
```

```python
df.info()
```

```python
df.dropna(subset=['Target date'], inplace=True)
```

```python
df.dropna(subset=['Baseline year'], inplace=True)
```

```python
df['Target date'] = df['Target date'].astype(float).astype(int)
```

```python
df["Baseline year"] = df["Baseline year"].replace("Other", 2005)
```

```python
df.info()
```

```python
df
```

```python
df1 = df[["Country", "City", "C40" , "Reporting Year", "Baseline year", "Baseline emissions","Percentage reduction target","Target date", "City Location", "Country Location"]]
```

```python
df1
```

```python
check_null(df1)
```

# 2017

```python
df_r17 = pd.read_csv("../DB/Data/2017_-_Cities_Emissions_Reduction_Targets_20240207.csv")
```

```python
df_r17.head()
```

```python
df_r17['C40'] = df_r17['C40'].apply(lambda x: convert_to_boolean(x))
```

```python
df_r17.rename(columns={'Baseline emissions (metric tonnes CO2e)': 'Baseline emissions'}, inplace=True)
df_r17.rename(columns={'Reporting year': 'Reporting Year'}, inplace=True)
```

```python
df_r17["Target date"] = df_r17["Target date"].replace("Other:", 2020)
```

```python
df_r17.dropna(subset=['Target date'], inplace=True)
```

```python
df_r17["Baseline year"] = df_r17["Baseline year"].replace("Other", 2005)
```

```python
df2 = df_r17[["Country", "City", "C40" , "Reporting Year", "Baseline year", "Baseline emissions","Percentage reduction target","Target date", "City Location", "Country Location"]]
```

```python
df2
```

```python
check_null(df2)
```

# Combine 2016 and 2017

```python
df_combined = pd.concat([df1, df2], ignore_index=True)
```

```python
df_combined["Target date"] = df_combined["Target date"].astype(str).str[-4:].astype(int)
```

```python
df_combined.dropna(subset=['Baseline year'], inplace=True)
df_combined["Baseline year"] = df_combined["Baseline year"].astype(str).str[-4:].astype(int)
```
