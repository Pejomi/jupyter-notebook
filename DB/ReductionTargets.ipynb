```python
import pandas as pd
```

```python
def check_null(df):
    # Check for null values in each column
    null_counts = df.isnull().sum()

    # Print columns with null values
    print("Columns with null values:")
    print(null_counts[null_counts > 0])
```

```python
# Function to convert C40 to True and NaN to False
def convert_to_boolean(value):
    if value == 'C40':
        return True
    else:
        return False
```

# 2016

```python
df = pd.read_csv("../DB/Data/2016_-_Cities_Emissions_Reduction_Targets_20240207.csv")
```

```python
df
```

```python
check_null(df)
```

```python
df['C40'] = df['C40'].apply(lambda x: convert_to_boolean(x))
```

```python
df.drop(columns=['Organisation'], inplace=True)
```

```python
df.rename(columns={'Baseline emissions (metric tonnes CO2e)': 'Baseline emissions'}, inplace=True)
```

```python
df.rename(columns={'City Name': 'City'}, inplace=True)
```

```python
df
```

```python
df1 = df[["Country", "City", "C40" , "Reporting Year", "Baseline year", "Baseline emissions","Percentage reduction target","Target date", "City Location", "Country Location"]]
```

```python
df1
```

```python
check_null(df1)
```
