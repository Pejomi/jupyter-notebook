#### Data Preparation for Training

```python
base_path = path
subdirs = ["Cat", "Dog"]
desired_size = (64, 64)

# Load and preprocess images
all_images = []
all_labels = []
counter = 0

print("Preprocessing images...")
for subdir in subdirs:
    image_paths = glob.glob(os.path.join(base_path, subdir, "*.jpg"))
    for image_path in image_paths:
        img = cv2.imread(image_path)
        if img is not None:
            counter += 1
            # Resize and normalize
            resized_img = cv2.resize(img, desired_size)
            normalized_img = resized_img / 255.0
            all_images.append(normalized_img)
            all_labels.append(subdir)
            print(subdir + ":", round((counter / len(image_paths)) * 100, 1), "%")
    counter = 0

# Convert images to NumPy arrays
all_images = np.array(all_images)

# Encode labels to numeric format
label_encoder = LabelEncoder()
encoded_labels = label_encoder.fit_transform(all_labels)

print("Saving data...")
np.save("./data/dogs-cats/preprocessed/images.npy", all_images)
np.save("./data/dogs-cats/preprocessed/labels.npy", encoded_labels)
```

#### Split data

```python
# Define the split ratios
train_ratio = 0.7
val_ratio = 0.15
# test ratio will be the remaining 15%

# Load the preprocessed images and labels
images = np.load("./data/dogs-cats/preprocessed/images.npy")
labels = np.load("./data/dogs-cats/preprocessed/labels.npy")

# Create a combined dataset to shuffle together
combined_data = list(zip(images, labels))

# Shuffle the combined dataset
np.random.shuffle(combined_data)

# Separate the combined dataset back into images and labels
shuffled_images, shuffled_labels = zip(*combined_data)

# Determine the split indices
total_samples = len(shuffled_images)

# Compute the indices for splitting
train_idx = int(total_samples * train_ratio)
val_idx = int(total_samples * (train_ratio + val_ratio))

# Split the data
train_images = shuffled_images[:train_idx]
train_labels = shuffled_labels[:train_idx]

val_images = shuffled_images[train_idx:val_idx]
val_labels = shuffled_labels[train_idx:val_idx]

test_images = shuffled_images[val_idx:]
test_labels = shuffled_labels[val_idx:]

# Save the training set
np.save("./data/dogs-cats/split-data/train_images.npy", np.array(train_images))
np.save("./data/dogs-cats/split-data/train_labels.npy", np.array(train_labels))

# Save the validation set
np.save("./data/dogs-cats/split-data/val_images.npy", np.array(val_images))
np.save("./data/dogs-cats/split-data/val_labels.npy", np.array(val_labels))

# Save the testing set
np.save("./data/dogs-cats/split-data/test_images.npy", np.array(test_images))
np.save("./data/dogs-cats/split-data/test_labels.npy", np.array(test_labels))
```

#### Load and prepare data

```python
# Load the saved data
train_images = np.load("./data/dogs-cats/split-data/train_images.npy")
train_labels = np.load("./data/dogs-cats/split-data/train_labels.npy")

val_images = np.load("./data/dogs-cats/split-data/val_images.npy")
val_labels = np.load("./data/dogs-cats/split-data/val_labels.npy")

test_images = np.load("./data/dogs-cats/split-data/test_images.npy")
test_labels = np.load("./data/dogs-cats/split-data/test_labels.npy")

# Check the shapes to confirm correct split
print("Train images shape:", train_images.shape)
print("Validation images shape:", val_images.shape)
print("Test images shape:", test_images.shape)

# Create a TensorFlow dataset from the loaded data
training_dataset = tf.data.Dataset.from_tensor_slices((train_images, train_labels))
training_dataset = training_dataset.shuffle(buffer_size=1024).batch(32)

validation_dataset = tf.data.Dataset.from_tensor_slices((val_images, val_labels))
validation_dataset = validation_dataset.shuffle(buffer_size=1024).batch(32)
```

#### Model Architecture

```python
    model = tf.keras.Sequential([
        layers.Conv2D(32, (3, 3), activation='relu', input_shape=(64, 64, 3)),
        layers.MaxPooling2D((2, 2)),
        layers.Conv2D(64, (3, 3), activation='relu'),
        layers.MaxPooling2D((2, 2)),
        layers.Conv2D(128, (3, 3), activation='relu'),
        layers.MaxPooling2D((2, 2)),
        layers.Flatten(),
        layers.Dense(128, activation='relu', kernel_regularizer=tf.keras.regularizers.L2(0.01)),
        layers.Dropout(0.5),
        layers.Dense(1, activation='sigmoid')
    ])
```

```python
model.summary()
```

```python
# Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

#### Model Training

```python
history = model.fit(training_dataset, epochs=10, validation_data=validation_dataset)
```

#### Model Evaluation

```python
# Evaluate the model on the test dataset
results = model.evaluate(test_images, test_labels)
print("Test Loss:", results[0])
print("Test Accuracy:", results[1])
```

```python
history_df = pd.DataFrame(history.history) 
history_df.loc[:, ['loss', 'val_loss']].plot() 
history_df.loc[:, ['accuracy', 'val_accuracy']].plot() 
plt.show() 
```

#### Model Testing and Prediction

```python
from tensorflow.keras.preprocessing import image
  
#Input image 
test_image = image.load_img('./data/dogs-cats/cat.jpg',target_size=(64,64)) 
  
#For show image 
plt.imshow(test_image) 
test_image = image.img_to_array(test_image) 
test_image = np.expand_dims(test_image,axis=0) 
  
# Result array 
result = model.predict(test_image) 

#Mapping result array with the main name list 
i=0
print("Prediction:", result[0][0])
if(result>=0.5): 
  print("Dog") 
else: 
  print("Cat")
```
